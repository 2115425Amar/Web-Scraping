```python
import csv
```

```python
import pandas as pd
```

```python
# Step 2: Scrape names and prices
names =  [name.text.strip() for name in soup.find_all("a", class_="title")]
prices = [price.text.strip() for price in soup.find_all("h4", class_="price")]
```

```python
csv_filename = "product_data.csv"
```

```python
with open(csv_filename, mode="w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["Product Name", "Price"])  # Writing headers
    writer.writerows(zip(names, prices))  # Writing data rows
```

```python

print(f"Data saved to {csv_filename} successfully.")
```

```python
# Step 4: Store the data in an Excel file
excel_filename = "product_data.xlsx"
data = {"Product Name": names, "Price": prices}
df = pd.DataFrame(data)
df.to_excel(excel_filename, index=False)
```
